```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
iris = sns.load_dataset('iris')
print(iris.head(10))
```

```python
sns.barplot(x='species', y='sepal_length', data=iris)
plt.title('Average Sepal Length for Each Species')
plt.show()
```

```python
sns.histplot(data=iris, x='sepal_length', hue='species', kde=True)
plt.title('Histogram of Sepal Length by Species')
plt.show()
```

```python
iris.groupby('species').sepal_length.mean().plot(kind='area')
plt.title('Area Plot of Sepal Length Mean by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length Mean')
plt.show()
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
plt.title('Sepal Length vs Petal Length by Species')
plt.show()
```

```python
iris.plot.hexbin(x='sepal_length', y='petal_length', gridsize=20, sharex=False, colormap='viridis')
plt.title('Hexbin Plot of Sepal Length vs Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

```python
iris['species'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Pie Chart of Species Distribution')
plt.ylabel('')
plt.show()
```
