```python
import pandas as pd
data = pd.read_csv('titanic.csv')
data.head()
```

```python
# Display the first few rows of the dataset
data.info()
data.describe()
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
missing_values
```

```python
# Fill missing values in 'Age' with the median age
data['age'].fillna(data['age'].median(), inplace=True)

# Fill missing values in 'Embarked' with the most common value
data['embarked'].fillna(data['embarked'].mode()[0], inplace=True)

# Drop the 'Cabin' column
data.drop(columns=['cabin'], inplace=True)

# Verify that there are no more missing values
data.isnull().sum()
```

```python
# Extract titles from names
data['Title'] = data['name'].str.extract(' ([A-Za-z]+)\.', expand=False)

# Replace rare titles with 'Rare' and standardize common titles
data['Title'] = data['Title'].replace(['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
data['Title'] = data['Title'].replace('Mlle', 'Miss')
data['Title'] = data['Title'].replace('Ms', 'Miss')
data['Title'] = data['Title'].replace('Mme', 'Mrs')

# Create a 'FamilySize' feature
data['FamilySize'] = data['sibsp'] + data['parch'] + 1

# Drop the 'Name' and 'Ticket' columns
data.drop(columns=['name', 'ticket'], inplace=True)
```

```python
# Encode categorical features
data = pd.get_dummies(data, columns=['sex', 'embarked', 'Title'], drop_first=True)

# Display the first few rows to verify the changes
data.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Select numerical features
numerical_features = ['age', 'fare', 'FamilySize']

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the numerical features
data[numerical_features] = scaler.fit_transform(data[numerical_features])

# Display the first few rows to verify the changes
data.head()
```
